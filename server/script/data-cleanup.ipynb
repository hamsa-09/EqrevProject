```python
import pandas as pd

file_path = "D:\Eqrev\server\data\\bquxjob_22a71535_199d6b590b1.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
# get row and column count
print(df.shape)
```

# Remove Columns with all NaN

```python
# Remove columns where all values are NaN
df_column_cleaned = df.dropna(axis=1, how='all')

df_column_cleaned.head()
```

```python
df_column_cleaned.shape
```

# Fill NaN Brand Names
Fill missing brand_name values by matching with existing non-null values.

```python
# Check current state of brand_name column
print("Brand name column info:")
print(f"Total rows: {len(df_column_cleaned)}")
print(f"Non-null brand_name values: {df_column_cleaned['brand_name'].notna().sum()}")
print(f"Null brand_name values: {df_column_cleaned['brand_name'].isna().sum()}")
print("\nUnique brand names (non-null):")
print(df_column_cleaned['brand_name'].dropna().unique())
```

```python
# Make an explicit copy
df_column_cleaned = df_column_cleaned.copy()

# Now fill NaN values
df_column_cleaned['brand_name'] = df_column_cleaned['brand_name'].ffill()  # forward fill
df_column_cleaned['brand_name'] = df_column_cleaned['brand_name'].bfill()  # backward fill (optional)


print(f"Null brand_name values after filling: {df_column_cleaned['brand_name'].isna().sum()}")
```

```python
df_column_cleaned.shape
```

```python
df_column_cleaned.head()
```

# Remove rows with NaN product id and product name

```python
# Drop rows where product_id or product_name is NaN
df_row_column_cleaned = df_column_cleaned.dropna(subset=['product_id', 'product_name'])

# Check the result
df_row_column_cleaned.shape
```

```python
df_row_column_cleaned.head()
```

```python
# Regex to match common date formats: YYYY-MM-DD, YYYY/MM/DD, DD/MM/YY, DD-MM-YY
date_pattern = r'^(\d{4}[-/]\d{2}[-/]\d{2})|(\d{2}[-/]\d{2}[-/]\d{2})$'

# Keep rows where product_id does NOT match the date pattern
df_cleaned = df_row_column_cleaned[~df_row_column_cleaned['product_id'].astype(str).str.match(date_pattern)]

df_cleaned.shape
```

# Fill zero to the for the remaining NaN values

```python
# Replace all NaN values in the DataFrame with 0
df_cleaned = df_cleaned.fillna(0)

# Check the result
df_cleaned.head()

# df_cleaned.shape
```

```python
df_cleaned.to_csv('cleaned_data.csv', index=False, encoding='utf-8')
```
